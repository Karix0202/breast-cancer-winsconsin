```python
import pandas as pd

df = pd.read_csv('datasets/data.csv')

df.head()
```

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
df.head()
```

```python
mining_set = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
mining_set['diagnosis'] = label_encoder.fit_transform(mining_set.values[:, 0])
```

```python
mining_set.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

mining_set.hist(bins=30, figsize=(20, 15))
```

```python
corr_matrix = mining_set.corr()['diagnosis'].sort_values(ascending=False)

corr_matrix
```

```python
import seaborn as sns
from matplotlib import rcParams

rcParams['figure.figsize'] = (20, 15)

heatmap = sns.heatmap(mining_set.corr())
```

```python
import numpy as np

cls_with_smallest_corr = mining_set[corr_matrix[np.abs(corr_matrix) < 0.1].keys()]

for i_name in cls_with_smallest_corr.columns:
    for j_name in cls_with_smallest_corr.columns:
        if i_name != j_name:
            mining_set[i_name + '/' + j_name] = mining_set[i_name] / mining_set[j_name]
```

```python
new_corr_matrix = mining_set.corr()['diagnosis'].sort_values(ascending=False)

new_corr_matrix
```

```python
for col_name in new_corr_matrix.keys():
    if len(col_name.split('/')) > 1 and abs(new_corr_matrix[col_name]) < 0.1:
        mining_set = mining_set.drop(col_name, axis=1)
        new_corr_matrix = new_corr_matrix.drop(labels=[col_name])
```

```python
mining_set.head()
```

```python
new_corr_matrix
```

```python
to_look = []

for key_name in new_corr_matrix.keys():
    splitted = key_name.split('/')
    if len(splitted) > 1:
        for j in splitted:
            to_look.append(j)
to_look = list(dict.fromkeys(to_look))

ids = {}
for col_name in to_look:
    ids[col_name] = df.columns.get_loc(col_name)
```

```python
ids
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class AttributesAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        fractal_dimension_se_per_smoothness_se = X[:, 20] / X[:, 15]
        fractal_dimension_se_per_symmetry_se = X[:, 20] / X[:, 19]
        fractal_dimension_se_per_fractal_dimension_mean = X[:, 20] / X[:, 10]
        smoothness_se_per_texture_se = X[:, 15] / X[:, 12]
        texture_se_per_fractal_dimension_se = X[:, 12] / X[:, 20]
        symmetry_se_per_fractal_dimension_se = X[:, 19] / X[:, 20]
        fractal_dimension_mean_per_fractal_dimension_se = X[:, 10] / X[:, 20]
        smoothness_se_per_fractal_dimension_se = X[:, 15] / X[:, 20]
        
        return np.c_[X,
                     fractal_dimension_se_per_smoothness_se,
                     fractal_dimension_se_per_symmetry_se,
                     fractal_dimension_se_per_fractal_dimension_mean,
                     smoothness_se_per_texture_se,
                     texture_se_per_fractal_dimension_se,
                     symmetry_se_per_fractal_dimension_se,
                     fractal_dimension_mean_per_fractal_dimension_se,
                     smoothness_se_per_fractal_dimension_se]
```

```python
from sklearn.model_selection import train_test_split

X = df.values[:, 1:32]
y = df.values[:, 0]
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('attrs_adder', AttributesAdder()),
    ('std_scaler', StandardScaler())
])

main_pipeline = FeatureUnion([
    ('num_pipeline', num_pipeline)
])
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_train_prepared = main_pipeline.fit_transform(X_train)

knn = KNeighborsClassifier()
knn.fit(X_train_prepared, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_train_predicted = knn.predict(X_train_prepared)
train_acc = accuracy_score(y_train, y_train_predicted)

X_test_prepared = main_pipeline.fit_transform(X_test)
y_test_predicted = knn.predict(X_test_prepared)
test_acc = accuracy_score(y_test, y_test_predicted)

'Train acc: {}, test acc: {}'.format(train_acc, test_acc)
```

```python
from sklearn.metrics import roc_auc_score

train_roc_score = roc_auc_score(y_train, y_train_predicted)
test_roc_score = roc_auc_score(y_test, y_test_predicted)

'Train roc auc score: {}, test roc auc score: {}'.format(train_roc_score, test_roc_score)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_test_predicted))
'\n'
print(classification_report(y_test, y_test_predicted))
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_test_predicted)
```

```python
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(knn, X_train_prepared, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_pred)

conf_mx
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train_prepared, y_train)
```

```python
y_train_predicted_svc = svc.predict(X_train_prepared)
y_test_predicted_svc = svc.predict(X_test_prepared)

train_svc_score = accuracy_score(y_train, y_train_predicted_svc)
test_svc_score = accuracy_score(y_test, y_test_predicted_svc)

'Train score: {}, test score {}'.format(train_svc_score, test_svc_score)
```

```python
train_roc_score_svc = roc_auc_score(y_train, y_train_predicted_svc)
test_roc_score_svc = roc_auc_score(y_test, y_test_predicted_svc)

'Train roc auc score: {}, test roc auc score: {}'.format(train_roc_score_svc, test_roc_score_svc)
```

```python
f1_score_train_svc = f1_score(y_train, y_train_predicted_svc)
f1_score_test_svc = f1_score(y_test, y_test_predicted_svc)

'Train f1 score: {}, test f1 score: {}'.format(f1_score_train_svc, f1_score_test_svc)
```

```python
from sklearn.model_selection import GridSearchCV

params = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
svc = SVC()
grid_search = GridSearchCV(svc, params, cv=5, scoring='accuracy')
grid_search.fit(X_train_prepared, y_train)
```

```python
final_model = grid_search.best_estimator_

y_final_pred_train = final_model.predict(X_train_prepared)
y_final_pred_test = final_model.predict(X_test_prepared)
```

```python
train_final_score = accuracy_score(y_train, y_final_pred_train)
test_final_score = accuracy_score(y_test, y_final_pred_test)

'Train score: {}, test score {}'.format(train_final_score, test_final_score)
```

```python
f1_score_train_final = f1_score(y_train, y_final_pred_train)
f1_score_test_final = f1_score(y_test, y_final_pred_test)

'Train f1 score: {}, test f1 score: {}'.format(f1_score_train_final, f1_score_test_final)
```
